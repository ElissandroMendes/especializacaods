Um POS TAG identificará a parte do discurso para uma sequência de palavras.

```
import nltk
nltk.download('gutenberg')
```

```
text = "I walked to the cafe to buy coffee before work."
```

```
text
```

Vamos inicialmente tokenizar o texto

```
nltk.download('punkt')
```

```
tokens = nltk.word_tokenize(text)
```

```
tokens
```

```
nltk.download('averaged_perceptron_tagger')
```

```
nltk.pos_tag(tokens)
```

Trabalhando com listas de palavras

```
md = nltk.corpus.gutenberg.words("melville-moby_dick.txt")
```

VAmos normalizar o texto

```
md_norm = [word.lower() for word in md if word.isalpha()]
```

```
md_tags = nltk.pos_tag(md_norm)
```

```
md_tags
```

```
md_tags[:5]
```

Vou querer criar uma lista apenas com os nomes contidos no texto

```
md_nouns = [word for word in md_tags if word[1] == "NN"]
```

```
md_nouns
```

```
len(md_nouns)
```

```
nouns_fd = nltk.FreqDist(md_nouns)
```

```
nouns_fd
```

```
nouns_fd.most_common()[:10]
```
