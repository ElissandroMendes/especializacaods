Tratamento de texto - CSV

```
import nltk
nltk.download('punkt')
import csv
```

Importando texto armazenado nas células em uma planilha ou em um arquivo csv (valores separados por vírgula).

```
comments = []
with open("reviews.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        comments.append(row)
```

Este comando nos dará uma lista, onde cada elemento é outra lista de todos os tokens para cada comentário.

```
tokens = [nltk.word_tokenize(str(entry)) for entry in comments]
```
