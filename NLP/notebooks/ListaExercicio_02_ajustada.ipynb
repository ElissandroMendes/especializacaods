```python
import nltk
nltk.download('gutenberg')
nltk.download('punkt')

%matplotlib inline
```

# Questão 1

```python
hamlet_books = ['shakespeare-caesar.txt','shakespeare-hamlet.txt','shakespeare-macbeth.txt']
hamlet_words_sents = {book : (nltk.corpus.gutenberg.words(book), nltk.corpus.gutenberg.sents(book)) for book in hamlet_books}
```

## 1.a - 1.e)

```python
for name, words_sents in hamlet_words_sents.items():
  print(f'Livro: {name}')  
  print(f'\tTotal palavras: {len(words_sents[0])}')
  print(f'\tTotal sentenças: {len(words_sents[1])}')
  print(f'\tTotal palavras não repetidas: {len(set(words_sents[0]))}')  
  print(f'\tTotal palavras repetidas: {len(words_sents[0]) - len(set(words_sents[0]))}')  
```

# Questão 2

## 2.a - 2.c)

```python
tam_maior = 0
tam_menor = 999999999
maior_doc = ''
menor_doc = ''
words_by_doc = []
total_words_corpus = 0
total_sents_corpus = 0
for document in nltk.corpus.gutenberg.fileids():
  words = nltk.corpus.gutenberg.words(document)
  sents = nltk.corpus.gutenberg.sents(document)
  len_words = len(words)
  len_sents = len(sents)  
  # Armazenando nro palavras e o documento para usar o max() e min()  
  words_by_doc.append((len_words, document))
  
  total_words_corpus += len_words
  total_sents_corpus += len_sents
    
  num_words = len(words)
  num_sents = len(sents)
  print(f'Total palavras para {document}: {num_words}')  
  print(f'Media palavras por sentença: {num_words / num_sents}')  
  print('\r')  
  if (num_words < tam_menor):
    tam_menor = num_words
    menor_doc = document

  if (num_words > tam_maior):
    tam_maior = num_words
    maior_doc = document
print('\r')
print(f'Maior documento: {maior_doc}')
print(f'Menor documento: {menor_doc}')
print('\r')
print(f'Maior documento MAX: {max(words_by_doc)}')
print(f'Menor documento MIN: {max(words_by_doc)}')
```

## 2.d )

```python
media_sents_por_words = total_sents_corpus / total_words_corpus
print(f'Média de sentenças por palavras no corpus: {media_sents_por_words}')
```

## 2.e

```python
macbeth_words = nltk.corpus.gutenberg.words('shakespeare-macbeth.txt')
macbeth_dist_freq = nltk.FreqDist(macbeth_words)

macbeth_dist_freq.plot(20, cumulative=False)
```

## 2.f)

```python
corpus_files = nltk.corpus.gutenberg.fileids()
corpus = nltk.corpus.gutenberg

most_common_5 = [dict(nltk.FreqDist(corpus.words(document)).most_common(5)) for document in corpus_files]

most_common_total_by_word = {} 
for d in most_common_5: 
    for k in d.keys(): 
        most_common_total_by_word[k] = most_common_total_by_word.get(k, 0) + d[k] 
        
ordered_result = sorted(most_common_total_by_word.items(), key = lambda kv:(kv[1], kv[0]))
print(ordered_result[-5:])
```
